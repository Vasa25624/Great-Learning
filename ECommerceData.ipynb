In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. The E-Commerce company has the following rule for computing their own revenue:

The company charges
(i) 25% on the orders having final price (discounted price) greater than 600
(ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600
(iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350
(iv) Otherwise, 5% on the final price (discounted price) </font>

```python
#function to determine the revenue
def compute_rev(x):
    if x > 600:
        return x*0.25
    elif x > 350:
        return x*0.15
    elif x > 100:
        return x*0.10
    else:
        return x*0.05

ECommerceData['Revenue'] = ECommerceData['discounted_price'].apply(compute_rev)
ECommerceData.head()
```

```python
# get the total revenue and print it
total_rev = ECommerceData['Revenue'].sum()
print('The total revenue generated by the ECom company across all orders is', total_rev, 'rupees')
```

```python
ECommerceData['BrandRevenue']= ECommerceData['discounted_price']-ECommerceData['Revenue'] 
ECommerceData.groupby(['Brand'])['BrandRevenue'].sum().sort_values(ascending = False).head(10)
```

```python
#Total number of orders placed per brand
NoOfOrders=ECommerceData['Brand'].value_counts()
print(NoOfOrders)
```

```python
#total retail_price per brand
RetailPriceBrandWise=ECommerceData.groupby(['Brand'])['retail_price'].sum()
print(RetailPriceBrandWise)
```

```python
# total discount_price per brand
DiscountPriceBrandWise=ECommerceData.groupby(['Brand'])['discounted_price'].sum()
print(DiscountPriceBrandWise)
```

```python
#total BrandRevenue generated per brand. 
BrandRevenueFinal=ECommerceData.groupby(['Brand'])['BrandRevenue'].sum().sort_values(ascending = False)
```

```python
DF_New=pd.DataFrame(columns=['RetailPriceBrandWise','NoOfOrders','DiscountPriceBrandWise', 'BrandRevenueFinal'])
DF_New['RetailPriceBrandWise']=RetailPriceBrandWise
DF_New['NoOfOrders']=NoOfOrders
DF_New['DiscountPriceBrandWise']=DiscountPriceBrandWise
DF_New['BrandRevenueFinal']=BrandRevenueFinal
DF_New
```

```python
plt.figure(figsize=(16,10))
sns.pairplot(data=DF_New[['BrandRevenueFinal','RetailPriceBrandWise','DiscountPriceBrandWise','NoOfOrders']])
```

```python
ECommerceData_Add = pd.DataFrame()
ECommerceData['Order_Date'] = pd.to_datetime(ECommerceData.Order_Date, format='%d-%m-%Y')
ECommerceData_Add['Order_Month']=ECommerceData['Order_Date'].dt.strftime("%Y %m")
ECommerceData_Add['Region']=ECommerceData['Region'].copy()
ECommerceData_Add['Revenue']=ECommerceData['Revenue'].copy()
ECommerceData_Add_New=pd.DataFrame(ECommerceData_Add.groupby(['Region','Order_Month'])['Revenue'].sum())
ECommerceData_Add_New
```

```python
plt.figure(figsize = (20,7))
#plt.xticks(np.arange(0, len('months')+7, 1))
#plt.xlim(1, 12)
ECommerceData_Add_New.sort_values(by='Order_Month', ascending=True, inplace=True)
plt.xticks(np.arange(0,14,1))
sns.lineplot(data=ECommerceData_Add_New,x = 'Order_Month' , y = 'Revenue', ci = False ,hue='Region');
plt.show()
```
