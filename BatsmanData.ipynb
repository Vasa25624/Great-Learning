```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
BatsmanData=pd.read_excel('D:/Ment Class/Practise11/Excel Data/BatsmanData.xlsx')
```

```python
# Count plot of Dismissals
sns.countplot(x ='Dismissal',
              data = BatsmanData,
              order = BatsmanData['Dismissal'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
# Countplot of Opposition - horizontally plotted
sns.countplot(y=BatsmanData['Opposition'],order = BatsmanData['Opposition'].value_counts().index)
plt.show()
```

```python
ct = pd.crosstab(BatsmanData['Opposition'],BatsmanData['Dismissal'], margins=True )
ct.drop('All',inplace = True)
ct.drop('All',axis = 1,inplace = True)
ct.plot.bar(stacked = True,figsize = (20,7))
plt.show()
```

```python
# Bar plot of Total runs scored against each Oppostion
sns.barplot( x='RunsScored', y='Opposition', estimator = np.sum, data=BatsmanData, ci=None)
# ci=None: Display of confidence intervals is not expected
#estimator = np.sum: Statistical function to estimate within each categorical bin sum wise
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(x= 'RunsScored', y='Opposition', data=BatsmanData)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot
sns.boxplot(x="RunsScored", y="post_transform", hue="Inning", data=BatsmanData, palette='Set2')

# Show the plot
plt.show()
```

```python
pd.crosstab(index = BatsmanData['post_transform'],columns=BatsmanData['Dismissal'], values=BatsmanData['RunsScored'],margins=True,aggfunc='mean')
```

```python
ct = pd.crosstab(index = BatsmanData['post_transform'],columns=BatsmanData['Dismissal'], values=BatsmanData['RunsScored'],margins=True,aggfunc='mean')
ct.drop('All',inplace = True)
ct.drop('All',axis = 1,inplace = True)
ct.plot.bar(figsize = (10,6))
plt.ylabel("Average Run Scored")
plt.show()
```
